```python
phase1_release = pd.read_table(
    "/kwiat/vector/ag1000g/release/phase1.AR3.1/samples/cross.samples.meta.txt",
    index_col=0)
```

```python
p2_df = phase2_release.query("cross == '29-2'").set_index("ox_code")
p1_df = phase1_release.query("cross == '29-2'")
```

```python
meta = p1_df[["cross", "role", "sex", "colony_id"]].join(
    p2_df[["n_reads",  "median_cov", "mean_cov"]])
```

```python
acc_stem = "ERZ{n}"
```

```python
meta["accession"] = [acc_stem.format(n=n) for n in range(407167, 407167 + len(meta))]
```

```python
ftp_path = 'ftp://ftp.sra.ebi.ac.uk/vol1/ERZ407/{ac}/{sid}.bam'
```

```python
meta["ftp_path"] = meta.apply(
    lambda y: ftp_path.format(
        ac=y.accession, sid=y.name.replace("-", "_")),
    axis=1)
```

```python
col_order = ["cross", "role", "n_reads", "median_cov", "mean_cov", 
             "sex", "colony_id", "accession", "ftp_path"]
meta[col_order].to_csv("../samples.csv")
```
